```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```python
DATADIR =r"C:\Users\kouadio\Robox\Dataset"
CATEGORIES =["Kaggle_condensateur", "Kaggle_led", "Kaggle_inductance"]

for category in CATEGORIES:
    path= os.path.join(DATADIR,category) #path to each category
    for img in os.listdir(path):
        img_array= cv2.imread(os.path.join(path,img), cv2.IMREAD_COLOR)
        #plt.imshow(img_array)
        #plt.show()
        
    
```

```python
IMG_SIZE =227 


new_array= cv2.resize(img_array, (IMG_SIZE,IMG_SIZE))
plt.imshow(new_array)
plt.show()
```

```python
training_data =[]

def create_training_data():
    for category in CATEGORIES:
        path= os.path.join(DATADIR,category) #path to each category
        print("The current path is: ", path)
        class_num = CATEGORIES.index(category)
        print("The class number is:  ",class_num)
        for img in os.listdir(path):
            img_array= cv2.imread(os.path.join(path,img), cv2.IMREAD_COLOR)
            new_array= cv2.resize(img_array, (IMG_SIZE,IMG_SIZE))
            training_data.append([new_array,class_num])            

create_training_data()
    
```

```python
print(len(training_data))
```

```python
#Shuffle the data before feeding it to the neural network

import random

random.shuffle(training_data)
```

```python
X_kaggle= []
y_kaggle= []
```

```python
for features, label in training_data:
    X_kaggle.append(features)
    y_kaggle.append(label)
```

```python
X_kaggle = np.array(X_kaggle).reshape(len(X_kaggle), IMG_SIZE, IMG_SIZE, 3)
```

```python
import pickle

pickle_out =open("X_kaggle.pickle","wb")
pickle.dump(X_kaggle,pickle_out)
pickle_out.close()

pickle_out =open("y_kaggle.pickle","wb")
pickle.dump(y_kaggle,pickle_out)
pickle_out.close()
```
